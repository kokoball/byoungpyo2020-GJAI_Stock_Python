```python
import pandas as pd
import numpy as np
import re
import csv
```

```python
# 파일을 불러온다
bin_list = []
f = open('data/종목별 핫이슈 키워드.csv', 'r', encoding='utf-8')
rdr = csv.reader(f)
for line in rdr:
  bin_list.append(line)
f.close()    

# 컬럼은 따로 저장
column = bin_list[0]

# 컬럼을 제외한 내용을 dataframe으로 저장
df = pd.DataFrame(bin_list[1:], columns=column)

#  1 부터 -1를 한 이유는 작은 따옴표 제거를 위해서!
print(df['ISSUE_CTNT'][0][1:-1])
```

```python
df.head()
```

```python
# 'ISSUE_CTNT'  에서 문자들만 추출하는 함수
def preprocess(str_list):
  p = re.compile("'.+?'")
  matched = p.findall(str_list)

  w_list = []
  for w in matched:
    kor = w.replace("'",'')
    if kor != 'words' and kor != 'wordweights':
      w_list.append(kor)
  # print(w_list)
  return w_list
```

```python
# data['ISSUE_CTNT']를 전부 전처리 해보자.

kor_data = []

for w in df['ISSUE_CTNT']:
  kor_list = preprocess(w[1:-1])
  kor_data.append(kor_list)

df['ISSUE_CTNT'] = kor_data

df.head()
```

```python
# 필요한 컬럼만 따로 저장
data = df[['ITMT_CD_NM','ISSUE_CTNT']]
```

### 빈도수별 종목 핫이슈 분류

```python
import collections
```

```python
delete_list = ['한국','단기','분기','주가','매수','매도','계약','공시','이익',
               '실적','차트','상승세','차트' ,'중기','순매도','투자','상장','배당','PBR','PER',
               '업종','지난해','쳐','최근','취득','체결','흑자','매출액','증가','규모','선','종목',
               '거래','추가','손실','사업','데일리','한경','배','티']
```

```python
def change_keyword(x):
    sample = data[data['ITMT_CD_NM']== x]
    
    a = collections.Counter()

    for i in range(len(sample)):
        a.update(sample.iloc[i,1])
        l = collections.Counter(a)
        xx = l.most_common()
        result = xx
    for i in xx:
        if len(i[0]) == 1 or x == i[0]:
            result.remove(i)
    for i in xx:
        if i[0] in delete_list:
            result.remove(i)
    return result[:7]
```

```python
change_keyword('라파스')
```

```python
name_list = list(set(data['ITMT_CD_NM']))
keyword_list = []

for i in range(len(name_list)):
    keyword_list.append(change_keyword(name_list[i]))
    
# df1 과 df2 를 하나로 합친다.
df1 = pd.DataFrame({'name': name_list})
df2 = pd.DataFrame({'words': keyword_list})
```
